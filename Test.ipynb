```python
%matplotlib inline 
from __future__ import division #
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(10, 0, 1000,endpoint=False) 
# 00(0,10]010
plt.ylim(ymin=0) # y 
plt.ylim(ymax=20) # y
# 2yy
for i in range(50):    
    y = 4*x**5/(i+1)**2+(i+1)/x #  
    """
    
    
    
    y = 4*x**5/(i+1)**2+(i+1)/(x+1)**3
    U
    U
    """
    ax.plot(x, y, 'k', linewidth=1, alpha = 1)
plt.show()
```
