```python
%matplotlib inline 
from __future__ import division #これを含まないと整数を整数で割った時の値がガウス記号つけたみたいになる
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(10, 0, 1000,endpoint=False) 
# 0を含むと平均費用の式に0で割られる部分が出るため(0,10]の範囲にするため0と10を入れ替えて終点を含まない、とした
plt.ylim(ymin=0) # グラフに表示するyの下限を定めている 
plt.ylim(ymax=20) # 同様にyの上限を定めている
# 上2行でyの範囲を指定してやらないとyの上限が際限なく上に伸びていってしまう
for i in range(50):    
    y = 4*x**5/(i+1)**2+(i+1)/x # 短期平均費用曲線の一例 
    """
    上のはどこかの大学の試験に出たらしい式をそのまま写してきたものですが、
    長期平均費用が見た感じ右上がりの直線になっててあまりよくないですね……。
    たとえば
    y = 4*x**5/(i+1)**2+(i+1)/(x+1)**3
    とやるともう少しそれっぽくなりますが、U字型にはなりません
    長期平均費用曲線がU字型になるような方程式を探してみるとよいでしょう
    """
    ax.plot(x, y, 'k', linewidth=1, alpha = 1)
plt.show()
```

```python
%matplotlib inline
from __future__ import division #これを含まないと整数を整数で割った時の値がガウス記号つけたみたいになる
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(10, 0, 1000,endpoint=False) 
# 0を含むと平均費用の式に0で割られる部分が出るため(0,10]の範囲にするため0と10を入れ替えて終点を含まない、とした
plt.ylim(ymin=0) # グラフに表示するyの下限を定めている 
plt.ylim(ymax=100) # 同様にyの上限を定めている
for i in range(30):    
    y = 4*x**6/(i+1)**2+(i+1) # 短期総費用曲線の一例  
    ax.plot(x, y, 'k', linewidth=1, alpha = 1)
plt.show()
```
