```python
%matplotlib inline
from __future__ import division
import matplotlib.pyplot as plt
import numpy as np
fig,ax=plt.subplots()
x=np.linspace(10,0.1000,endpoint=False)
plt.ylim(ymin=0)
plt.ylim(ymax=200)
for i in range(50):
    y=4*x**9/(i+1)**2+(i+1) #短期費用関数
    ax.plot(x,y,'k',linewidth=1,alpha=1)
plt.show()
```

```python
%matplotlib inline
from __future__ import division
import matplotlib.pyplot as plt
import numpy as np
fig,ax=plt.subplots()
x=np.linspace(10,0,1000,endpoint=False) # 変更点1
"""
引数の0.1000（ピリオド）を0,1000（カンマ）に変更。
変更前はlinspaceの中の引数が10,0.1000の２つになっていた。
変更後は10,0,1000の３つ。
linspaceは決められた範囲をいくつかの区間に分けて区間ごとに直線を引いていく関数。
引数を３つ取るときは１つ目が書き始めの点、２つ目が書き終わりの点、３つ目が分ける区間の数となる。
引数を２つ取る際は１つ目が書き始めの点、２つ目が書き終わりの点、そして区間の数は自動的に50となる。
(下の2つのページがわかりやすいかも)
http://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html
http://www.maroon.dti.ne.jp/koten-kairo/works/fft/Section_PythonGraph/python_graph5.html
このため、修正前だと範囲が0.1から10までになってしまっていた。
また、この影響で区間内に打たれる点が50個になる。50個だとこのグラフのような大きい傾きを表すのには
ちょっと粗すぎるかもしれない。
"""
plt.ylim(ymin=0)
plt.ylim(ymax=200)
for i in range(50):
    y=4*x**9/(i+1)**2+(i+1) #短期費用関数
    ax.plot(x,y,'y',linewidth=0.5,alpha=0.8) #変更点2
"""
ここに関しては変える必然性はまったくないのですが……
ax.plotは3～5番目の引数によって線の色や太さ、濃さを変えることができます。
色に関しては下のURLを見てください。
http://matplotlib.org/api/colors_api.html
太さや濃さはそのままです。linewidthが線の太さ、alphaは線の濃さを表します。
太さはどんな値を入れてもいいですが濃さは0～1の間におさめてください。
引く曲線の本数も含めて適当に調整し、
包絡線の様子がいちばん見やすくなるような値を入れるとよいと思います。
"""
plt.show()
```

```python
%matplotlib inline
from __future__ import division
import matplotlib.pyplot as plt
import numpy as np
fig,ax=plt.subplots()
x=np.linspace(10,0.1000,endpoint=False)
plt.ylim(ymin=0)
plt.ylim(ymax=200)
for i in range(20):
    y=4*x**8/(i+1)**2+(i+1)/x #短期平均費用曲線
    
    """長期平均費用曲線が９X^２になるように考えてみました
汎用性を持たせるための改変部のまとめ？は生産関数などの仮定した数字を一般式にしたものを
書けばよいのでしょうか？よくわかっていません。。。"""
    
    ax.plot(x,y,'k',linewidth=1,alpha=1)
plt.show()

# x=L**a*K**b　生産関数　L=労働投入量　K=資本投入量
#短期費用曲線　wL+rK w=賃金率　r=レンタル率
#短期平均費用曲線　y=w*L/x+r*K/x




```
